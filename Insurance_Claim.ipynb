```python
cat_cols=data.select_dtypes(exclude=[np.number]).columns.tolist()
```

```python
cat_cols
```

```python
fig,axes=plt.subplots(nrows=len(cat_cols),ncols=1,figsize=(10,6*len(cat_cols)))
for i,column in enumerate(cat_cols):
    grouped=data.groupby(column)['Claim_Amount'].mean().sort_values()
    axes[i].bar(grouped.index,grouped.values)
plt.tight_layout()
```

```python
fig,axes=plt.subplots(nrows=len(cat_cols),ncols=1,figsize=(10,6*len(cat_cols)))
for i,column in enumerate(cat_cols):
    sns.countplot(x=column,data=data,ax=axes[i],palette='Set2')
plt.tight_layout()
```

```python
grouped_data=data.groupby(cat_cols)['Claim_Amount'].agg(['mean','count']).reset_index()
high_risk=grouped_data[grouped_data['mean']>grouped_data['mean'].quantile(0.75)]
```
